```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/ayuatikahhh/Preprocessing-leaderboard-history/refs/heads/main/dataset.csv')
df
```

```python
df.info()
```

terdapat
5 baris pada kolom team_nema yang null dan
4 baris pada kolom elo yang null
nama kolom dan elo itu hal penting untuk analisis jadi karena itu sebaiknya kita hilangkan saja baris yang tidak lengkap informasinya

```python
df = df.dropna(subset=["team_name", "elo"])
df.reset_index(drop=True, inplace=True)
df
```

```python
df.info()
```

timestampe data type nya berupa objek kita perlu ubah menjadi datetime

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df
```

```python
df.info()
```

```python
dupes = df[df.duplicated()]
print(dupes)
```
